### Construct the proper sentences from individual words and print the 5 sentences.

```python
def createSentences(path):
    train_sentences = []
    temp = []
    with open(path, "r") as f:
        for word in f:
            if word == '\n':
                train_sentences.append(temp)
                temp = []
            else:
                temp.append(word.rstrip('\n'))
            
    f.close()
    return train_sentences       
```

## Defining features for CRF





```python
# Let's define the features to get the feature value for one word.
# Functinos to get the POS of a word

def most_common(lst):
    return max(set(lst), key=lst.count)

def get_POS(word):
    try:
        return most_common(list(pos_map.loc[pos_map['Words']==word, 'POS'].values))
    except:
        return 'NONE'
```

```python
# Check POS of word 'second'

get_POS('second')
```

```python
# Function to get CRF features when a split sentence is passed with the position of the number
def getFeaturesForOneWord(sentence, pos, tags):
           
      
    word = sentence[pos]
    tag = tags[pos]

    features = [
        'word.lower=' + word.lower(),                                           # Word ID
        'word[-3:]=' + word[-3:],                                               # Last Three Characters
        'word[-2:]=' + word[-2:],                                               # Last Two Characters
        'word.isupper=%s' % word.isupper(),                                     # is word in all Uppercase
        'word.isdigit=%s' % word.isdigit(),                                     # is word a Number
        'words.startsWithCapital=%s' % word[0].isupper(),                       # does word start with a Capital letter
        'word.POS=' + tag                                                       # POS of word
      ]

    if(pos > 0):                                              # Select features of previous word
        prev_word = sentence[pos-1]
        prev_tag = tags[pos-1]
        features.extend([
        'prev_word.lower=' + prev_word.lower(), 
        'prev_word.isupper=%s' % prev_word.isupper(),
        'prev_word.isdigit=%s' % prev_word.isdigit(),
        'prev_words.startsWithCapital=%s' % prev_word[0].isupper(),
        'prev_word.POS=' + prev_tag
      ])
    else:
        features.append('BEG')                                  # Tag beginning word

    if(pos == len(sentence)-1):
        features.append('END')                                  # Tag ending word

    return features


# Function to pass the sentence words along with their tags with position of word, iterating through each word
def getFeaturesForOneSentence(sentence):
    tags = []

    for i in sentence:
        tags.append(get_POS(i))
    
    return [getFeaturesForOneWord(sentence, pos, tags) for pos in range(len(sentence))]
```

## Define input and target variables


Correctly computing X and Y sequence matrices for training and test data.
Check that both sentences and labels are processed

### Define the features' values for each sentence as input variable  for CRF model in test and the train dataset 

```python
start_time = time.monotonic()

X_train = [getFeaturesForOneSentence(sentence) for sentence in train_sentences]
y_train = [labels for labels in train_labels]
 
X_test = [getFeaturesForOneSentence(sentence) for sentence in test_sentences]
y_test = [labels for labels in test_labels]

end_time = time.monotonic()
print(timedelta(seconds=end_time - start_time))
```

### Define the labels as the target variable for test and the train dataset

## Build the CRF Model

```python
# Build the CRF model.
crf = sklearn_crfsuite.CRF(max_iterations=250)

try:
    crf.fit(X_train, y_train)
except AttributeError:                          # Exception handling required on my PC. May not be necessary on other Systems
    pass
```

## Evaluation

### Predict the labels of each of the tokens in each sentence of the test dataset that has been pre processed earlier.

```python
y_pred = crf.predict(X_test)
metrics.flat_f1_score(y_test, y_pred, average='weighted')
```

##Identifying Diseases and Treatments using Custom NER

We now use the CRF model's prediction to prepare a record of diseases identified in the corpus and treatments used for the diseases.



### Create the logic to get all the predicted treatments (T) labels corresponding to each disease (D) label in the test dataset.

image.png

```python
# Getting only the sentences which has both 'D' and 'T' label. The others are not useful for our Evaluation

final_sentences = []
final_preds = []

for sentence, prediction in zip(test_sentences, y_pred):
    if 'D' in prediction:
        if 'T' in prediction:
            final_sentences.append(sentence)
            final_preds.append(prediction)
```

```python
print(' '.join(final_sentences[0]))
print(' '.join(final_preds[0]))
```

### Predict the treatment for the disease name: 'hereditary retinoblastoma'

```python
#Extracting the Disease name and the Treatement

disease_pattern = 'D+'
treat_pattern = 'T+'

disease = []
treatment = []

for sentence, prediction in zip(final_sentences, final_preds):    
    temp_disease = []
    temp_treatment = []
    
    for match in re.finditer(disease_pattern, ''.join(prediction)):
        temp = []
        for i in range(match.start(), match.end()):
            temp.append(sentence[i])
        # print(' '.join(temp))
        temp_disease.append(temp)
        

    for match in re.finditer(treat_pattern, ''.join(prediction)):
        temp = []
        for i in range(match.start(), match.end()):
            temp.append(sentence[i])
        # print(' '.join(temp))
        temp_treatment.append(temp)
        
    disease.append(temp_disease)
    treatment.append(temp_treatment)
        
    print(' '.join(sentence))
    print()
        
    for i in temp_disease:
        print('disease  -->  ', end='')
        print(' '.join(i), end=', ')
        print()
    
    for i in temp_treatment:
        print('treatment  -->  ', end='')
        print(' '.join(i), end=', ')
        print()
    
    print('---------------------------------------------------------------')
```

```python
#Check if length of diseases and treatment is the same

print(len(disease))
print(len(treatment))
```
